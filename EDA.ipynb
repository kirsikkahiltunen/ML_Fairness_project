```python
df = df.drop(["kunta_nimi", "vuosi", "vuosikuukausi"], axis=1)
df.shape
```

```python
df["ka_korvatut_paivat"] = df["korvattu_paiva_lkm"].div(df["saaja_lkm"])
df["ka_maksettu_eur"] = df["maksettu_eur"].div(df["korvattu_paiva_lkm"])
```

```python
print("ka € min:", df["ka_maksettu_eur"].min())
print("ka € max:", df["ka_maksettu_eur"].max())
print("ka € median:", df["ka_maksettu_eur"].median())

print("ka pv min:", df["ka_korvatut_paivat"].min())
print("ka pv max", df["ka_korvatut_paivat"].max())
print("ka_pv_median", df["ka_korvatut_paivat"].median())
```

```python
categorical_cols = ["aikatyyppi", "ikaryhma", "sukupuoli", "etuus", "korvausperuste", "kunta_nro"]
df_encoded = pd.get_dummies(df, columns=categorical_cols, drop_first=False, dtype=int)
print(df_encoded.shape)
df_encoded.head()
```

```python
df_encoded = df_encoded.iloc[:300000]
```

### Train models

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import HistGradientBoostingRegressor
gc.collect()
X = df_encoded.drop(["ka_korvatut_paivat", "ka_maksettu_eur"], axis=1)
gc.collect()
y_days = df_encoded["ka_korvatut_paivat"]
gc.collect()
y_eur = df_encoded["ka_maksettu_eur"]
gc.collect()

training_indices = int(len(X)*0.8)
X_train = X.iloc[:training_indices]
X_test = X.iloc[training_indices:]
y_train_days = y_days.iloc[:training_indices]
y_test_days = y_days.iloc[training_indices:]

y_train_eur = y_eur.iloc[:training_indices]
y_test_eur = y_eur.iloc[training_indices:]

lr_days = LinearRegression().fit(X_train, y_train_days)
hgb_days = HistGradientBoostingRegressor(max_bins=128).fit(X_train, y_train_days)
lr_eur = LinearRegression().fit(X_train, y_train_eur)
hgb_eur = HistGradientBoostingRegressor(max_bins=128).fit(X_train, y_train_eur)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import numpy as np

def evaluate_model(y_true, y_pred, model_name="model"):
    mse = mean_squared_error(y_true, y_pred)
    rmse = np.sqrt(mse)
    mae = mean_absolute_error(y_true, y_pred)
    r2 = r2_score(y_true, y_pred)
    
    print(f"{model_name}")
    print(f"R^2:\t{r2:.4f}")
    print(f"MAE:\t{mae:.4f}")
    print(f"MSE:\t{mse:.4f}")
    print(f"RMSE:\t{rmse:.4f}\n")
```

```python
pred_days_lr = lr_days.predict(X_test)
pred_days_hgb = hgb_days.predict(X_test)
pred_eur_lr = lr_eur.predict(X_test)
pred_eur_hgb = hgb_eur.predict(X_test)

eval_days_lr = evaluate_model(y_test_days, pred_days_lr, "LinearRegression (Days)")
eval_days_hgb = evaluate_model(y_test_days, pred_days_hgb, "HistGradientBoosting (Days)")

eval_eur_lr = evaluate_model(y_test_eur, pred_eur_lr, "LinearRegression (EUR)")
eval_eur_hgb = evaluate_model(y_test_eur, pred_eur_hgb, "HistGradientBoosting (EUR)")
```

### Testing fairness

```python
def fairness_by_group(y_true, y_pred, X_test, sensitive_prefix):
    sens_cols = [col for col in X_test.columns if col.startswith(sensitive_prefix)]

    results = {}

    for col in sens_cols:
        mask = X_test[col] == 1

        group_name = col.replace(sensitive_prefix, "").replace("_", "")

        group_y_true = y_true[mask]
        group_y_pred = y_pred[mask]

        results[group_name] = {
            "count": mask.sum(),
            "mean_pred": group_y_pred.mean(),
            "mean_true": group_y_true.mean(),
            "mean_error": (group_y_pred - group_y_true).mean(),
            "MAE": mean_absolute_error(group_y_true, group_y_pred),
            "RMSE": mean_squared_error(group_y_true, group_y_pred) ** 0.5
        }

    return pd.DataFrame(results).T
```

```python
y_pred_days = lr_days.predict(X_test)

fairness_days = fairness_by_group(
    y_true = y_test_days,
    y_pred = y_pred_days,
    X_test = X_test,
    sensitive_prefix="sukupuoli_"
)

print("Fairness assessment for Days by sex (LR):")
print(fairness_days)

fairness_days = fairness_by_group(
    y_true = y_test_days,
    y_pred = y_pred_days,
    X_test = X_test,
    sensitive_prefix="ikaryhma_"
)

print("\nFairness assessment for Days by age group (LR):")
print(fairness_days)
```

```python
y_pred_days = hgb_days.predict(X_test)

fairness_days = fairness_by_group(
    y_true = y_test_days,
    y_pred = y_pred_days,
    X_test = X_test,
    sensitive_prefix="sukupuoli_"
)

print("Fairness assessment for Days by sex (HGB):")
print(fairness_days)

fairness_days = fairness_by_group(
    y_true = y_test_days,
    y_pred = y_pred_days,
    X_test = X_test,
    sensitive_prefix="ikaryhma_"
)

print("\nFairness assessment for Days by age group (HGB):")
print(fairness_days)
```

```python
y_pred_eur = lr_eur.predict(X_test)

fairness_eur = fairness_by_group(
    y_true = y_test_eur,
    y_pred = y_pred_eur,
    X_test = X_test,
    sensitive_prefix="sukupuoli_"
)

print("Fairness assessment for EUR by sex (LR):")
print(fairness_eur)

fairness_eur = fairness_by_group(
    y_true = y_test_eur,
    y_pred = y_pred_eur,
    X_test = X_test,
    sensitive_prefix="ikaryhma_"
)

print("\nFairness assessment for EUR by age group (LR):")
print(fairness_eur)
```

## Fairness-aware predictive model

### Inverse probability weighting for age-group

```python
import gc
gc.collect()
age_groups = df["ikaryhma"].value_counts(normalize=True)
df["weights"] = df["ikaryhma"].map(lambda x: 1/ age_groups[x])
df["weights"] = ((df["weights"])/df["weights"].mean())
df.head()
```

```python
categorical_cols = ["aikatyyppi", "ikaryhma", "sukupuoli", "etuus", "korvausperuste", "kunta_nro"]
df_encoded2 = pd.get_dummies(df, columns=categorical_cols, drop_first=False, dtype=int)
print(df_encoded2.shape)
df_encoded2.head()
```

```python
df_encoded2 = df_encoded2.iloc[:300000]
```

### Train models

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import HistGradientBoostingRegressor
```

```python
gc.collect()
X = df_encoded2.drop(["ka_korvatut_paivat", "ka_maksettu_eur"], axis=1)
gc.collect()
y_days = df_encoded2["ka_korvatut_paivat"]
gc.collect()
y_eur = df_encoded2["ka_maksettu_eur"]
```

```python
training_indices = int(len(X)*0.8)
X_train = X.iloc[:training_indices]
X_test = X.iloc[training_indices:]
y_train_days = y_days.iloc[:training_indices]
y_test_days = y_days.iloc[training_indices:]
```

```python
training_indices = int(len(X)*0.8)
X_train = X.iloc[:training_indices]
X_test = X.iloc[training_indices:]
y_train_eur = y_eur.iloc[:training_indices]
y_test_eur = y_eur.iloc[training_indices:]
```

```python
lr_days = LinearRegression().fit(X_train, y_train_days, sample_weight=X_train["weights"])
```

```python
hgb_days = HistGradientBoostingRegressor(max_bins=128).fit(X_train, y_train_days, sample_weight=X_train["weights"])
```

```python
lr_eur = LinearRegression().fit(X_train, y_train_eur, sample_weight=X_train["weights"])
```
